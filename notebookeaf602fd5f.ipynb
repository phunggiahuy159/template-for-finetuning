```python
!pip install sentence_transformers
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader
import pandas as pd 
import wandb
wandb.login(key="67ef10d66b47e95e78d3a3286bc9b6a93382a0cb")

# Load pre-trained model
model = SentenceTransformer("/kaggle/input/finall/output/riel_output2")

# Create triplet examples
csv_file = "/kaggle/input/final-triplet/data_for_sentence_transformer.csv"  # Replace with your actual file path
data = pd.read_csv(csv_file).dropna()

# Convert the CSV data into InputExample format
train_examples = [
    InputExample(texts=[row['query'], row['positive_passage'], row['negative_passage_1']])
    for _, row in data.iterrows()
]

# Convert to DataLoader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=32)

# Define triplet loss
train_loss = losses.TripletLoss(model=model)

# Fine-tune the model
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=4,
    output_path="output/riel_output3"
)
wandb.finish()

```
